```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X,y =make_regression(n_samples=100,n_features=1,noise=15,random_state=42) #Generate random 100 sample with nooise for realism
m=X.shape[0]
y=y.reshape(-1,1)
print(X.shape)
print(y.shape)
```

```python
X_b=np.c_[np.ones((m,1)),X]# add columns and value = 1 for intercept term
```

```python
theta=np.array([[2.0],[3.0]])# initialize the parameter intercept and slope
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, X_b.dot(theta), color="green", label="Initial Line (No GD)") # This performs matrix multiplication.and predict the value of y
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression Without Gradient Descent")
plt.legend()
plt.show()
```

```python
learning_rate=0.1
n_iteration=100
i=1
for _ in range(n_iteration):
    y_pred=X_b.dot(theta)
   
    gradients = (2 / m) * X_b.T.dot(y_pred - y) # calculate Gradient
 
    theta -= learning_rate * gradients
    print(theta)

```

```python
plt.figure(figsize=(10, 5))
print(theta)
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, X_b.dot(theta), color="red", label="Optimized Line (With GD)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression With Gradient Descent")
plt.legend()
plt.show()
```
